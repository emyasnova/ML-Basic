## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
def change_case(text: str, req_case_type: str = None) -> str:
    is_pascal_case = text[0].isupper()
    is_pascal_case_req = req_case_type == 'PascalCase' or not (req_case_type or is_pascal_case)
    
    if (is_pascal_case == is_pascal_case_req):
        return text

    res = ''
    skip_next = False
    for i, letter in enumerate(text):
        if is_pascal_case_req and letter == '_' :
            res += text[i+1].upper()
            skip_next = True
        elif is_pascal_case_req and i == 0:
            res += text[i].upper()
        elif not is_pascal_case_req and letter.isupper():
            if i != 0:
                res += '_'
            res += letter.lower()
        elif skip_next == True:
            skip_next = False
        else:
            res += letter
    return res

print(change_case('PythonIsTheBest', req_case_type='PascalCase'))
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def validate_date(date: str) -> bool:
    date, month, year = map(int, date.split('.'))
    if month > 12:
        return False
    if month in [1, 3, 5, 7, 8, 10, 12]:
        return date == 31
    if month in [4, 6, 9, 11]:
        return date == 30
    if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
        return date == 29
    else:
        return date == 28
    

print(validate_date('31.04.1962'))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def is_simple_number(number: int) -> bool:
    res = True
    for i in range(2, number - 1):
        if number % i == 0:
            res = False
            break
    return res

print(is_simple_number(17))
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def print_table(col_names: list[str], dict):
    print('\t'.join(col_names))
    for user in dict:
        print(f'{'\t'.join(dict[user])}\t{user}')
```

```python
users = {}
fields_names = ['name', 'surname', 'age', 'id']
i = 0

name, surname, age, id = (None, None, None, None)

while (True):
    string = input(fields_names[i] + ':')
    if string == '':
        break

    if i == 0:
        name = string
        if not name.isalpha():
            print('имя должно содержать только буквы')
            continue
        name = name[0].upper() + name[1:]
    elif i == 1:
        surname = string
        if not surname.isalpha():
            print('фамилия должна содержать только буквы')
            continue
        surname = surname[0].upper() + surname[1:]
    elif i == 2:
        age = string
        if (not age.isdigit() or int(age) < 18 or int(age) > 60):
            print('возраст должен быть целым числом от 18 до 60')
            continue
    elif i == 3:
        id = string
        if (not id.isdigit() or len(id) != 8):
            print('id должен быть целым числом, дополненным до 8 знаков незначащими нулями')
            continue
        if id in users:
            print('id должен быть уникальным')
            continue
        users[id] = (name, surname, age)
    i = 0 if i == 3 else i + 1

print_table(fields_names, users)
```
