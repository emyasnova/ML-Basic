## 5. Практическое задание

### 5.1. Создание пакета для анализа простых данных

Создайте пакет `simple_analytics` со следующей структурой:

```
simple_analytics/
├── __init__.py
├── basic_stats.py
├── data_loader.py
└── report_generator.py
```

**Задание:**

1. **basic_stats.py** - создайте функции:
   - `calculate_basic_stats(numbers)` - возвращает словарь с count, sum, mean, min, max
   - `find_outliers(numbers)` - находит выбросы (числа, сильно отличающиеся от среднего)

2. **data_loader.py** - создайте функции:
   - `load_from_text(filename)` - загружает числа из текстового файла
   - `load_from_csv(filename, column_name)` - загружает числа из определенного столбца CSV

3. **report_generator.py** - создайте функцию:
   - `generate_report(numbers, title="Отчет")` - создает текстовый отчет со статистикой

4. **__init__.py** - настройте импорты и `__all__`
