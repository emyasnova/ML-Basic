## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
string = input('Введите пятизначное число:')

new_string = string[0]
for i in range(3, 0, -1):
    new_string += string[i]
new_string += string[4]

print(new_string)
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
days = int(input('Введите количество дней до отпуска:'))

last_week_days = days % 7;
days_off = (days // 7) * 2
if (last_week_days > 5):
    days_off += last_week_days - 5


print(f'Количество выходных дней до отпуска: {days_off}')
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
##### bar_length = int(input('Введите длину плитки шоколада'))
bar_width = int(input('Введите ширину плитки шоколада'))
piece_size = int(input('Введите размер куска, который хотите отломить'))

print(piece_size % bar_length == 0 and piece_size // bar_length <= bar_width 
      or piece_size % bar_width == 0 and piece_size // bar_width <= bar_length)
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
digits_data = [
    {'one': 'I', 'five': 'V', 'ten': 'X', 'k': 1},
    {'one': 'X', 'five': 'L', 'ten': 'C', 'k': 10},
    {'one': 'C', 'five': 'D', 'ten': 'M', 'k': 100}
]

number = int(input('Целое положительное число до 4000:'))

result = ''

for data in digits_data:
    a = number % (data['k'] * 10) // data['k']
    number -= a * data['k']
    if (a == 5 - 1):
        result = data['one'] + data['five'] + result
    elif (a == 9):
        result = data['one'] + data['ten'] + result
    else:
        result = data['one'] * (a % 5) + result
        result = data['five'] * (a // 5) + result

print(result)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
string = input('Введите данные:')

index = string.find('.')
res = True
if (index != -1 and (string.count('.') > 1 or index == len(string))):
    res = False

print(res)
```
